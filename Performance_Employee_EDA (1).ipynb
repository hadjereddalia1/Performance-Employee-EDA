```python
# Define the function to count numerical and categorical columns
def count_data_types(df):
    # Identify numerical and categorical columns
    numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
    categorical_cols = df.select_dtypes(include=['object']).columns

    # Print the columns
    print("Numerical Columns: ", numerical_cols)
    print("Categorical Columns: ", categorical_cols)

    # Count the number of columns in each type
    num_numerical = len(numerical_cols)
    num_categorical = len(categorical_cols)

    # Print the count of each type
    print(f"Number of Numerical Columns: {num_numerical}")
    print(f"Number of Categorical Columns: {num_categorical}")

    return num_numerical, num_categorical

# Call the function to count data types and print the results
count_data_types(data)
```

```python
from scipy.stats import f_oneway

# List of numerical features
numerical_col = data.select_dtypes(include=['int64', 'float64']).columns

# Ensure PerformanceRating has multiple categories
ratings = data['PerformanceRating'].unique()

if len(ratings) > 1:
    for column in numerical_col:
        # Create groups for each PerformanceRating category
        groups = [data[data['PerformanceRating'] == rating][column] for rating in ratings]

        # Perform ANOVA test
        stat, p_value = f_oneway(*groups)

        # Print results
        print(f"Feature: {column}")
        print(f"F-statistic: {stat:.4f}, p-value: {p_value:.4e}")
        if p_value < 0.05:
            print("Significant difference between groups")
        else:
            print("No significant difference between groups")
        print("-" * 50)
else:
    print("PerformanceRating must have more than one category for ANOVA.")

```
